## 1> zip

```python
lis1 = [1,2,3]
lis2 = ["a", "b", "c"]
lis3 = "ABC"
print(zip(lis1,lis2,lis3))
print(list(zip(lis1,lis2,lis3)))
print(dict(zip(lis2,lis1)))
```

## 2> unzip

use * to unzip

```python
lis1 = [1,2,3]
lis2 = ["a", "b", "c"]
lis3 = "ABC"
li_zip = list(zip(lis1,lis2,lis3))
print(li_zip)
print(*li_zip, sep ="\n")
```

```python
print(*li_zip)
```

```python
print(list(zip(*li_zip)))
print(list(zip((1, 'a', 'A'), (2, 'b', 'B'), (3, 'c', 'C'))))
```

## Exercise 1

For a dictionary m = {'a': 1, 'b': 2, 'c': 3}, you should have a new dictionary m_reverse = {1: 'a', 2: 'b', 3: 'c'}

```python
m = {'a': 1, 'b': 2, 'c': 3}
m_reverse = {}
for key ,value in m.items():
    m_reverse[value] = key    
print(m_reverse)    
```

```python
m = {'a': 1, 'b': 2, 'c': 3}
m_reverse = dict(zip(m.values(), m.keys()))
print(m_reverse)
```

## 3> Filter

***filter(function, iterable)***   where we can use   **lambda**   function as the function, t will keep all elements that is **True**, when applying the function and keep the original order. The following example shows how to filter the prime number of a list

```python
import math
def check_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(math.sqrt(n)+1)):
        if  n % i == 0:
            return False
    return True

nums = range(100)
print(list(filter(check_prime, nums)))
```

## 4> Map

***map(function, iterable)*** is similar to filter. but it apply the function for all elements and return a new iterable.

```python
nums = range(100)
print(list(map(lambda x: x**2, nums)))
```

## 5> reduce
**`reduce(funciton_of_2_elements, iterable).`**
*e.g.* li = [1, 2, 3, 4], then `reduce(f,li)` will comput the result of `f(f(f(1,2),3),4)`
To use **`reduce`**, `from functools import reduce`

```python
from functools import reduce
nums = range(1,100)
print(reduce(lambda x,y: x+y, nums))
```

# Exercise 2: implement map and reduce

1. `map_(function, iterable)`, return an *iterator*
2. `reduce_(function, iterable)`, return the result

Hint: `yield`

```python
def map_(function, iterable):
    for i in iterable:
        yield function(i)
        
print(list(map(lambda x: x**2, range(10))))
print(list(map_(lambda x: x**2, range(10))))
```

```python
def reduce_(function, iterable):
    result = iterable[0]
    for i in iterable[1:]:
        result = function(result, i)
    return result

print(reduce(lambda x,y: x+y, range(10)))
print(reduce_(lambda x,y: x+y, range(10)))
```

```python
def mapreduce(func1, func2, iterable):
    li = map(func1, iterable)
    return reduce(func2, li)
```
